```python
# L1 parameters:number of set:32  block size:8  set associativity:1

import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Fixed L2 set associativity:1-way

x1=[32,64,128,256,512]
y1=[0.303,0.2913,0.2613,0.2254,0.1997]
x2=[32,64,128,256,512]
y2=[0.1775,0.1564,0.1322,0.1147,0.1119]
x3=[32,64,128,256,512]
y3=[0.1041,0.0874,0.0706,0.0686,0.0612]
x4=[32,64,128,256,512]
y4=[0.0605,0.0456,0.0441,0.0378,0.0374]
x5=[32,64,128,256,512]
y5=[0.0343,0.0316,0.025,0.0246,0.0246]
x6=[32,64,128,256,512]
y6=[0.0211,0.01560,0.0152,0.0152,0.0152]

l1=plt.plot(x1,y1,'r--',label='L2 block size:32')
l2=plt.plot(x2,y2,'g--',label='L2 block size:64')
l3=plt.plot(x3,y3,'b--',label='L2 block size:128')
l4=plt.plot(x4,y4,'k--',label='L2 block size:256')
l5=plt.plot(x5,y5,'c--',label='L2 block size:512')
l6=plt.plot(x6,y6,'m--',label='L2 block size:1024')

plt.plot(x1,y1,'ro-',x2,y2,'go-',x3,y3,'bo-',x4,y4,'ko-',x5,y5,'co-',x6,y6,'mo-')

plt.xlabel('The number of sets in L2')
plt.ylabel('Miss rate')

plt.title('L1 block size:8 bytes and Fixed L2 set associativity:1-way')
plt.legend()
plt.savefig('E:\\CA_plot\\fig_1_anagram.png')
plt.show()
```
